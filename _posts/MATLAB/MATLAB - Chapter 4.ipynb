# Question 4
The height and speed of a projectile (such as a thrown ball) launched with a
speed of $v_o$ at an angle $A$ to the horizontal are given by 

>$h(t) = v_o tsinA - 0.5gt^{2}$ \\
>$v(t) = \sqrt{v_o^{2} - 2v_ogtsinA + g^{2}t^{2}}$

where ð‘” is the acceleration due to gravity. The projectile will strike the ground
when h(t) = 0, which gives the time to hit $t_{hit} = 2(v_o/g)sinA$. Suppose that
$A = 40Â°$, $ð‘£_o = 20 ð‘š/ð‘ $, an
d ð‘” = 9.81 $ð‘š/ð‘ ^{2}$.
Use the MATLAB relational and logical operators to find the first time when $â„Ž\ge 6$.

```
import numpy as np
from matplotlib.pyplot import *

# initial values
A = np.deg2rad(40) #rad
g = 9.81 #m/s^2
vo = 20 #m/s

# 
t_hit = np.sqrt((2)*(vo/g)*(sin(A)))
t = np.arange(0,t_hit,0.01)
h = ((vo)*(t)*(np.sin(A))) - ((0.5)*(g)*(t**2))
v = np.sqrt( (vo**2) -((2)*(vo)*(g)*(t)*(np.sin(A))) + ((g**2)*(t**2)) )

h6 = np.where(h >= 6) # looks for indices with h>=6
th = t[h6] # creates array with times when h>=6
print(th[0]) # prints first instance

# plots
subplot(1,2,1)
plot(t,h)
title('h(t)')
subplot(1,2,2)
plot(t,v)
title('v(t)')
```

# Question 6
Create a function to evaluate the function $f(x,y)$ defined as follows:

>\begin{equation}
  f(x,y)  =
    \begin{cases}
      xy & \text{if $x \ge 0$ and $y \ge 0$}\\
      xy^{2} & \text{if $x \ge 0$ and $y \le 0$}\\
      x^{2}y & \text{if $x < 0$ and $y \ge 0$}\\
      x^{2}y^{2} & \text{if $x<0$ and $y<0$}
    \end{cases}       
\end{equation}

For x = -5, y =7, the value returned by the function f(x,y) is

```
def f(x,y):
  if (x>= 0 and y>=0):
    f = lambda x,y: x*y
    print(f(x,y))
  elif (x>=0 and y<0):
    f = lambda x,y: x*(y**2)
    print(f(x,y))
  elif (x<0 and y>=0):
    f = lambda x,y: (x**2)*y
    print(f(x,y))
  else:
    f = lambda x,y: (x**2)*(y**2)
    print(f(x,y))

f(-5,7)
```

# Question 8
Write a script to compute the sum of the first 15 terms in the series $5k^{2}-2k$ for $k=1,2,3,4,...,15$.

```
sum = 0
for k in range(1,16):
  print(k)
  sum = sum + (5*(k**2) - 2*k)
  k= k + 1

print(sum)
```
